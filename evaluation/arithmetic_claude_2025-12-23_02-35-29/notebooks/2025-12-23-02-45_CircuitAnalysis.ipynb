# Replication: Vector Arithmetic in Concept and Token Subspaces

## Overview
This notebook replicates the experiments from the arithmetic_eval repository, which demonstrates that concept and token induction heads can identify subspaces of Llama-2-7b activations with coherent semantic and surface-level structure, enabling more accurate parallelogram arithmetic (e.g., Athens  Greece + China = Beijing) than using raw hidden states.

## Key Components:
1. **OV Lens Construction**: Build concept and token lenses by summing OV matrices from top-k induction heads
2. **Word Embedding Extraction**: Pass words through Llama-2-7b and transform using lens matrices
3. **Parallelogram Arithmetic**: Compute La_  Lb_ + Lb'_ and measure nearest neighbor accuracy
4. **Comparison Settings**: raw (L=Id), concept lens, token lens, and all heads baseline

```python
# Import necessary libraries
import os
import json
import torch
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt
from typing import Dict, List, Tuple, Optional

# Set device
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"Using device: {device}")

# Set paths
REPO_PATH = '/net/scratch2/smallyan/arithmetic_eval'
DATA_PATH = os.path.join(REPO_PATH, 'data')
CACHE_PATH = os.path.join(REPO_PATH, 'cache')
REPLICATION_PATH = os.path.join(REPO_PATH, 'evaluation', 'replications')
```

## Step 1: Load Model and Head Scores

We load Llama-2-7b using nnsight for intervention-friendly access to hidden states.

```python
# Load the model using nnsight
from nnsight import LanguageModel

MODEL_NAME = 'meta-llama/Llama-2-7b-hf'
print(f"Loading model: {MODEL_NAME}")
model = LanguageModel(MODEL_NAME, device_map='cuda', dispatch=True)
print(f"Model loaded successfully")
print(f"Hidden size: {model.config.hidden_size}")
print(f"Number of layers: {model.config.num_hidden_layers}")
print(f"Number of attention heads: {model.config.num_attention_heads}")
```

```python
# Load concept and token head scores
def load_head_scores(head_type: str, model_name: str = 'Llama-2-7b-hf') -> List[Dict]:
    """Load head scores from cache."""
    score_path = os.path.join(CACHE_PATH, 'causal_scores', model_name, f'{head_type}_copying_len30_n1024.json')
    with open(score_path, 'r') as f:
        scores = json.load(f)
    return scores

# Load concept and token scores
concept_scores = load_head_scores('concept')
token_scores = load_head_scores('token')

# Sort by score descending and get top-k heads
def get_top_k_heads(scores: List[Dict], k: int = 80) -> List[Tuple[int, int]]:
    """Get top-k heads sorted by score."""
    sorted_scores = sorted(scores, key=lambda x: x['score'], reverse=True)
    return [(d['layer'], d['head_idx']) for d in sorted_scores[:k]]

TOP_K = 80
top_concept_heads = get_top_k_heads(concept_scores, TOP_K)
top_token_heads = get_top_k_heads(token_scores, TOP_K)

print(f"Top {TOP_K} concept heads (first 10):")
print(top_concept_heads[:10])
print(f"\nTop {TOP_K} token heads (first 10):")
print(top_token_heads[:10])
```

## Step 2: Build OV Lens Matrices

The OV lens is constructed by summing the OV matrices (O @ V) from top-k induction heads. This creates a transformation matrix that projects hidden states into a subspace that emphasizes either semantic (concept) or surface-level (token) information.

```python
def build_ov_lens(model, heads_to_sum: List[Tuple[int, int]], rank: int = 4096) -> torch.Tensor:
    """
    Build OV lens by summing OV matrices from specified heads.
    
    Args:
        model: The language model
        heads_to_sum: List of (layer, head_idx) tuples
        rank: Low-rank approximation (4096 = full rank for Llama-2-7b)
    
    Returns:
        ov_sum: (hidden_size, hidden_size) transformation matrix
    """
    hidden_size = model.config.hidden_size
    head_dim = hidden_size // model.config.num_attention_heads
    
    with torch.no_grad():
        ov_sum = torch.zeros((hidden_size, hidden_size), device='cuda')
        
        # Group heads by layer for efficient processing
        layers_to_process = set([l for l, _ in heads_to_sum])
        
        for layer in layers_to_process:
            heads_in_layer = [h for l, h in heads_to_sum if l == layer]
            for h in heads_in_layer:
                # V: (head_dim, hidden_size) - projects hidden state down
                V = model.model.layers[layer].self_attn.v_proj.weight[h * head_dim : (h+1) * head_dim]
                # O: (hidden_size, head_dim) - projects value back up
                O = model.model.layers[layer].self_attn.o_proj.weight[:, h * head_dim : (h+1) * head_dim]
                # OV: (hidden_size, hidden_size)
                ov_sum += torch.matmul(O, V)
        
        # Apply low-rank approximation if requested
        if rank < hidden_size:
            U, S, Vh = torch.linalg.svd(ov_sum)
            ov_sum = (U[:, :rank] * S[:rank]) @ Vh[:rank]
    
    return ov_sum

# Build all heads lens (for baseline comparison)
def build_all_heads_lens(model, rank: int = 4096) -> torch.Tensor:
    """Build lens using all attention heads."""
    all_heads = [(l, h) for l in range(model.config.num_hidden_layers) 
                 for h in range(model.config.num_attention_heads)]
    return build_ov_lens(model, all_heads, rank)

print("Building OV lenses...")
concept_lens = build_ov_lens(model, top_concept_heads)
token_lens = build_ov_lens(model, top_token_heads)
all_lens = build_all_heads_lens(model)

print(f"Concept lens shape: {concept_lens.shape}")
print(f"Token lens shape: {token_lens.shape}")
print(f"All heads lens shape: {all_lens.shape}")
```

## Step 3: Extract Word Representations

For each word, we:
1. Pass it through the model
2. Extract the hidden state at a specified layer
3. Optionally apply the OV lens transformation

```python
def get_word_representation(word: str, model, layer: int, ov_lens: Optional[torch.Tensor] = None, 
                            prefix: str = '', offset: int = -1) -> torch.Tensor:
    """
    Get word representation from the model at a given layer.
    
    Args:
        word: The word to encode
        model: The language model
        layer: Layer index to extract hidden state from
        ov_lens: Optional OV lens matrix to apply (None for raw hidden state)
        prefix: Optional prefix to add before the word
        offset: Token position to extract (-1 for last token)
    
    Returns:
        Word representation tensor
    """
    # Add prefix (with space if word doesn't start with special char)
    text = prefix + word.strip()
    
    with torch.no_grad():
        with model.trace(text):
            state = model.model.layers[layer].output[0].squeeze()[offset].detach().save()
    
    if ov_lens is not None:
        return torch.matmul(ov_lens, state)
    return state

def get_all_word_representations(words: set, model, layer: int, ov_lens: Optional[torch.Tensor] = None,
                                  prefix: str = '') -> Dict[str, torch.Tensor]:
    """Get representations for all words in a set."""
    representations = {}
    for word in words:
        representations[word] = get_word_representation(word, model, layer, ov_lens, prefix)
    return representations

# Test on a few sample words
test_words = {'Athens', 'Greece', 'Beijing', 'China'}
test_reps = get_all_word_representations(test_words, model, layer=20, ov_lens=concept_lens)
print("Sample word representations (concept lens at layer 20):")
for word, rep in test_reps.items():
    print(f"  {word}: shape={rep.shape}, norm={rep.norm().item():.4f}")
```

## Step 4: Parallelogram Arithmetic Evaluation

For each analogy (a : b :: a' : b'), we compute:
- Vector arithmetic: a - b + b' 
- Find the nearest neighbor among all candidate words
- Check if the nearest neighbor is a' (the correct answer)

```python
def compute_parallelogram_accuracy(task_lines: List[str], word_reps: Dict[str, torch.Tensor], 
                                    separator: str = ' ') -> Tuple[float, int]:
    """
    Compute nearest-neighbor accuracy for parallelogram arithmetic.
    
    For each analogy (a, b, a', b'): compute (a - b + b') and check if a' is nearest neighbor.
    
    Args:
        task_lines: Lines from task file, each with 4 words
        word_reps: Dictionary of word -> representation
        separator: Separator between words in task file
    
    Returns:
        accuracy: Fraction of correct nearest-neighbor predictions
        n: Number of examples
    """
    correct = 0
    total = 0
    
    for line in task_lines:
        parts = line.split(separator)
        if len(parts) != 4:
            continue
        
        a, b, a_prime, b_prime = parts
        
        # Skip if any word is missing
        if not all(w in word_reps for w in [a, b, a_prime, b_prime]):
            continue
        
        # Compute parallelogram: a - b + b' should equal a'
        result_vec = word_reps[a] - word_reps[b] + word_reps[b_prime]
        
        # Find nearest neighbor using cosine similarity
        best_similarity = -float('inf')
        best_word = None
        
        for word, rep in word_reps.items():
            sim = torch.cosine_similarity(result_vec, rep, dim=0).item()
            if sim > best_similarity:
                best_similarity = sim
                best_word = word
        
        if best_word == a_prime:
            correct += 1
        total += 1
    
    accuracy = correct / total if total > 0 else 0.0
    return accuracy, total

# Test on capital-common-countries
with open(os.path.join(DATA_PATH, 'word2vec', 'capital-common-countries.txt'), 'r') as f:
    lines = f.read().strip().split('\n')
# Skip header line
task_lines = [l for l in lines[1:] if l]

# Get all unique words
all_words = set()
for line in task_lines:
    parts = line.split(' ')
    if len(parts) == 4:
        all_words.update(parts)

print(f"Task: capital-common-countries")
print(f"Number of examples: {len(task_lines)}")
print(f"Unique words: {len(all_words)}")

# Quick test with concept lens at layer 20
test_reps = get_all_word_representations(all_words, model, layer=20, ov_lens=concept_lens)
test_acc, test_n = compute_parallelogram_accuracy(task_lines, test_reps)
print(f"\nQuick test - Concept lens at layer 20:")
print(f"Accuracy: {test_acc:.4f} ({int(test_acc * test_n)}/{test_n})")
```

## Step 5: Full Evaluation Across Layers and Lens Types

We evaluate parallelogram accuracy across:
- Multiple layers (0, 4, 8, 12, 16, 20, 24, 28, 31)
- Four lens types: raw, concept, token, all

```python
def evaluate_task(task_name: str, dataset: str, model, layers: List[int],
                  concept_lens: torch.Tensor, token_lens: torch.Tensor, all_lens: torch.Tensor,
                  prefix: str = '') -> Dict:
    """
    Evaluate parallelogram accuracy for a task across layers and lens types.
    """
    # Load task data
    sep = ' ' if dataset == 'word2vec' else '\t'
    task_path = os.path.join(DATA_PATH, dataset, f'{task_name}.txt')
    
    with open(task_path, 'r') as f:
        lines = f.read().strip().split('\n')
    task_lines = [l for l in lines[1:] if l]  # Skip header
    
    # Get all unique words
    all_words = set()
    for line in task_lines:
        parts = line.split(sep)
        if len(parts) == 4:
            all_words.update(parts)
    
    results = {
        'task': task_name,
        'n_examples': len(task_lines),
        'n_words': len(all_words),
        'layers': layers,
        'accuracies': {
            'raw': [],
            'concept': [],
            'token': [],
            'all': []
        }
    }
    
    lenses = {
        'raw': None,
        'concept': concept_lens,
        'token': token_lens,
        'all': all_lens
    }
    
    for layer in tqdm(layers, desc=f'{task_name}'):
        for lens_type, lens in lenses.items():
            word_reps = get_all_word_representations(all_words, model, layer, lens, prefix)
            acc, _ = compute_parallelogram_accuracy(task_lines, word_reps, sep)
            results['accuracies'][lens_type].append(acc)
    
    return results

# Evaluate on key tasks
LAYERS = [0, 4, 8, 12, 16, 20, 24, 28, 31]

# Start with capital-common-countries (semantic task)
print("Evaluating capital-common-countries...")
capital_results = evaluate_task('capital-common-countries', 'word2vec', model, LAYERS,
                                 concept_lens, token_lens, all_lens)
```
